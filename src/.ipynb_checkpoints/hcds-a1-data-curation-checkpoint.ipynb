# A1 - Data curation

## Step 1: Data acquisition

We start by importing the necessary python libraries for http requests handling.

````python
import json
import requests
````

Define strings for the Wikimedia REST API for legacy and pageviews endpoints.

>We use the {param_name} synthax here to plug in parameters using expanded dictionaries.

````python
endpoint_legacy = 'https://wikimedia.org/api/rest_v1/metrics/legacy/pagecounts/aggregate/{project}/{access-site}/{granularity}/{start}/{end}'
endpoint_pageviews = 'https://wikimedia.org/api/rest_v1/metrics/pageviews/aggregate/{project}/{access}/{agent}/{granularity}/{start}/{end}'
````

Define a function to execute http GET requests to the api endpoints.

````python
def api_call(endpoint, parameters, headers):
    call = requests.get(endpoint.format(**parameters), headers=headers)
    response = call.json()
    return response
````

We import date, datetime and relativedelta to help us in date manipulations in the next step. We want to be able to set dates in one place and have the rest of our analysis automatically updated.

````python
from datetime import date, datetime
from dateutil.relativedelta import *
````

We first create date objects that we will reuse during our analysis.

>Note we define the `end_month_base_params_date` date one month later. This is qwerk from the legacy api. Both apis use the same synthax to specify monthly intervals.

We then define query parameters for the GET requests that will remain constant. (`base_params`)

A few constants are defined to ease the dynamic creation of dictionaries for file naming

The 'headers' object is provided to abide to the global rules of usage for the api. It allows Wikimedia to contact us in case of misuse or other inqueries.

````python
# Define dates
# Parsing strptime logic reference: https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
api_date_format = "%Y%m%d%H"
start_month_date = datetime.strptime("2008010100", api_date_format)
end_month_date = datetime.strptime("2019080100", api_date_format)
end_month_base_params_date = end_month_date + relativedelta(months=+1)

# Define base params for api calls
base_params = {"project" : "en.wikipedia.org",
                 "granularity" : "monthly",
                 "start" : datetime.strftime(start_month_date, api_date_format),
                 "end" : datetime.strftime(end_month_base_params_date, api_date_format)
                }

# File naming constants
file_name_date_format = "%Y%m"
first_month = datetime.strftime(start_month_date, file_name_date_format)
last_month = datetime.strftime(end_month_date, file_name_date_format)

# Terms of use required headers
headers = {
    'User-Agent': 'https://github.com/ALotOfData',
    'From': 'bebrodeu@uw.edu'
}
````

The api exposes the following access types to the site, we will be creating a JSON file for each access type below using the `file_name_format` string.

> Note we use a relative file path to place the resulting files in the raw_data folder

````python
access_types_legacy = ["desktop-site", "mobile-site"]
access_types_pageviews = ["desktop", "mobile-app", "mobile-web"]
file_name_format = "../raw_data/{api_name}_{access_type}_{first_month}-{last_month}.json"
````

The method below is used to generate a json file per access_type given an api_name.

We set specific values, for instance the access parameter is named differently in the legacy and the pageviews api.

Note that the pageviews's api also has an `agent` query parameter which specifies to only get pageviews from users (excludes spiders and bots). This parameter is not available in the legacy api.

We use the ** synthax to unravel dictionaries into named parameters. We use this technique twice:
* to add the "access-type" value in the query parameters for the api call
* for the formatting of the file name.

````python
def create_json_files(api_name):
    if(api_name == "legacy"):
        endpoint = endpoint_legacy
        params = base_params
        access_types = access_types_legacy
        access_type_parameter_name = "access-site"
    # pageviews
    else:
        endpoint = endpoint_pageviews
        params = {**base_params, "agent": "user"}
        access_types = access_types_pageviews
        access_type_parameter_name = "access"
    
    file_format_params = {"api_name": api_name,
                          "first_month": first_month,
                          "last_month": last_month
                         }
    
    for access_type in access_types:
        complete_params = {**params, access_type_parameter_name: access_type}
        file_content = api_call(endpoint, complete_params, headers)
        file_name = file_name_format.format(**{**file_format_params, "access_type": access_type})
        with open(file_name, 'w') as file_to_write:
            json.dump(file_content, file_to_write)
````

We call the above function to create legacy and pageviews raw json files.

````python
create_json_files("legacy")
create_json_files("pageviews")
````

## Step 2: Data processing

We start by defining a function which indexes the raw json data based on date using the YYYY-MM format. The function will return dictionary of the form:

```json
{"date": view_count}
```

>Note if the filename contains 'legacy' we get the view count from the count property. The legacy and pageviews apis have different names for the count value.

````python
def get_month_indexed_dictionary(file_name):
    views_per_month = {}
    
    if 'legacy' in file_name:
        view_count_property_name = 'count'
    else: 
        view_count_property_name = 'views'
    
    with open(file_name) as file:
        content = json.load(file)
        for entry in content['items']:
            entry_date = datetime.strptime(entry['timestamp'], api_date_format)
            views = entry[view_count_property_name]
            views_per_month[entry_date.strftime('%Y-%m')] = views
        return views_per_month
````

We manually specify the names of the different raw json files to load and feed them to the function defined above to obtained date indexed dictionaries. We use the access_types conventions of the apis to identify the dictionaries.

````python
file_format_params = {"first_month": first_month, "last_month": last_month}

file_name = file_name_format.format(**{**file_format_params, "api_name": "legacy", "access_type": "desktop-site"})
desktop_site = get_month_indexed_dictionary(file_name)

file_name = file_name_format.format(**{**file_format_params, "api_name": "legacy", "access_type": "mobile-site"})
mobile_site = get_month_indexed_dictionary(file_name)

file_name = file_name_format.format(**{**file_format_params, "api_name": "pageviews", "access_type": "desktop"})
desktop = get_month_indexed_dictionary(file_name)

file_name = file_name_format.format(**{**file_format_params, "api_name": "pageviews", "access_type": "mobile-app"})
mobile_app = get_month_indexed_dictionary(file_name)

file_name = file_name_format.format(**{**file_format_params, "api_name": "pageviews", "access_type": "mobile-web"})
mobile_web = get_month_indexed_dictionary(file_name)
````

We create two functions:

* a first to iterate through all the months between two dates.
* a second to generate a row of data given a date
  1. We first set all counts to 0, it will be the default value if an api has no data for that month.
  2. We then use the dictionaries to check if data is availble. If data is available it is set.
  3. We create aggregate values for all views and mobile for both apis.

````python
# Modified version of: https://stackoverflow.com/questions/35066588/is-there-a-simple-way-to-increment-a-datetime-object-one-month-in-python/35067328
def get_month_range_inclusive(start, end):
    current = start
    while current <= end:
        yield current;
        current = current + relativedelta(months=+1)

def create_row(month_date):
    year = month_date.strftime('%Y')
    month = month_date.strftime('%m')
    month_index = month_date.strftime("%Y-%m")
    
    # legacy
    desktop_site_views = 0
    mobile_site_views = 0
    
    # pageviews
    desktop_views = 0
    mobile_app_views = 0
    mobile_web_views = 0
    
    if month_index in desktop_site.keys():
        desktop_site_views = desktop_site[month_index]
        
    if month_index in mobile_site.keys():
        mobile_site_views = mobile_site[month_index]
        
    if month_index in desktop.keys():
        desktop_views = desktop[month_index]
        
    if month_index in mobile_app.keys():
        mobile_app_views = mobile_app[month_index]
    
    if month_index in mobile_web.keys():
        mobile_web_views = mobile_web[month_index]
    
    # Aggregate values
    pagecount_all_views = desktop_site_views + mobile_site_views
    pageview_mobile_views = mobile_app_views + mobile_web_views
    pageview_all_views = desktop_views + pageview_mobile_views
        
    return [year, month, pagecount_all_views, desktop_site_views, mobile_site_views, pageview_all_views, desktop_views, pageview_mobile_views]
````

We use the two functions above to generate a 2d array containing one row for every month between our start and end dates.

````python
cleaned_data_array = []
for month_date in get_month_range_inclusive(start_month_date, end_month_date):
    row = create_row(month_date)
    cleaned_data_array.append(row)
````

We import and use the pandas library to generate a dataframes with specified column names.

We use a relative path to write the dataframe without indexes in a clean_data folder.

````python
import pandas as pd

clean_df = pd.DataFrame(cleaned_data_array, columns=['year', 'month', 'pagecount_all_views', 'pagecount_desktop_views', 'pagecount_mobile_views', 'pageview_all_views', 'pageview_desktop_views', 'pageview_mobile_views'])
cleaned_data_file_name = "../clean_data/en-wikipedia-traffic_{first_month}-{last_month}.csv".format(**{'first_month':first_month, 'last_month': last_month})
clean_df.to_csv(cleaned_data_file_name, index=None)
````

## Step 3: Analysis

We first import libraries needed to correctly format our graph and enable dataframe's numpy array's manipulations.

````python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
import matplotlib.ticker as ticker
import numpy as np
````

We read our cleaned data file and make sure that year and month are treated as strings.

We then remove the last monthly data point 2016-08 for the pagecount api as it is an incomplete entry. We replace these entries by 0.

We use pandas to transform our year and month columns into a new column called 'date' which has the np.datetime64 datatype

````python
clean_df = pd.read_csv(cleaned_data_file_name, dtype={'year': str, 'month': str}, header=0)
clean_df.loc[(clean_df.year == '2016') & (clean_df.month == '08'), ['pagecount_desktop_views', 'pagecount_all_views', 'pagecount_mobile_views']] = [0,0,0]
clean_df['date'] = pd.to_datetime(clean_df.year + clean_df.month, format="%Y%m")
````

We create a new data frame which changes unavailable data represented by 0s with nan which will not be plotted by matplotlib.

We remove year and month columns as we no longer need them.

We obtain page counts in millions as it will make our graph's axes cleaner.

````python
# nan is not plotted
plot_df = clean_df.replace(0, np.nan)

# no longer needed
plot_df = plot_df.drop(['year', 'month'], axis=1)

# Change to millions of view counts for an easier to read y axis
plot_df['pagecount_desktop_views'] /= 1e6
plot_df['pageview_desktop_views'] /= 1e6
plot_df['pagecount_all_views'] /= 1e6
plot_df['pageview_all_views'] /= 1e6
plot_df['pagecount_mobile_views'] /= 1e6
plot_df['pageview_mobile_views'] /= 1e6
````

The matplotlib gallery post below was very useful in understand the next few manipulations: https://matplotlib.org/3.1.1/gallery/text_labels_and_annotations/date.html

Some of its content was used directly below.

The specific formating steps are documented below.

````python
# Obtain max and min dates for x axis limits
# We add an extra year to the max date
datemin = np.datetime64(plot_df['date'][0], 'Y')
datemax = np.datetime64(plot_df['date'][plot_df.shape[0]-1], 'Y') + np.timedelta64(1, 'Y')


fig, ax = plt.subplots()

# Title and subscript text
ax.set_title('Monthly page views on English Wikipedia (x 1,000,000)', fontsize=16)
plt.suptitle("Solid lines mark pageviews api, which removes robots from viewcount.", y=0.05, color='red', fontsize=16);

# Set plot size
fig.set_size_inches(18, 6)

# Set plot lines
# We set _no_legend_ when we don't want a certain line to be part of the legend
ax.plot('date', 'pagecount_desktop_views', data=plot_df, linestyle='--', color='green', label='main site')
ax.plot('date', 'pageview_desktop_views', data=plot_df, linestyle='-', color='green', label='_nolegend_')
ax.plot('date', 'pagecount_all_views', data=plot_df, linestyle='--', color='black', label='total')
ax.plot('date', 'pageview_all_views', data=plot_df, linestyle='-', color='black', label='_nolegend_')
ax.plot('date', 'pagecount_mobile_views', data=plot_df, linestyle='--', color='blue', label='mobile site')
ax.plot('date', 'pageview_mobile_views', data=plot_df, linestyle='-', color='blue', label='_nolegend_')

# X label and limits
ax.set_xlabel('date')
ax.set_xlim(datemin, datemax)

# X axis ticks formating
years = mdates.YearLocator()   # every year
years_fmt = mdates.DateFormatter('%Y')
ax.xaxis.set_major_locator(years)
ax.xaxis.set_major_formatter(years_fmt)

# Y axis label
ax.set_ylabel('views (x 1,000,000)')

# Legend and grid on
ax.legend(loc='upper left', fontsize=16)
ax.grid(True)

# Save figure to png in the assets/image folder
plt.savefig('../assets/images/wikipedia_page_views.png')
plt.show()
````
